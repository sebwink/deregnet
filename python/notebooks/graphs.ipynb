The algorithms implemented in the deregnet package require a base graph as input. In principle you could use any directed graph. The package represents graphs as instances of igraph graphs and also provides some predefined graphs with biological semantics, like graphs constructed from KEGG pathway maps, etc.

```python
import igraph as ig
import deregnet.graphs  # this is the deregnet graphs module containing specific graphs
```

Ideally you now read the documentation of this module:

```python
help(deregnet.graphs)
```

As you can see, there are basically five graphs dats sources integrated into deregnet: 'KEGG', 'OmniPath', 'PathwayCommons', 'ReactomeFI' and 'RegNetwork'. You can find further information on each of these graphs (represented as classes of common base DeregnetGraph) by consulting the documentation:

```python
help(deregnet.graphs.RegNetwork)
```

Taking the 'RegNetwork' graph resource as an example, you can instantiate a graph ready for use with the deregnet algorithms, by creating an instance of the RegNetwork class:

```python
graph = deregnet.graphs.RegNetwork('hsa')
```

```python
graph
```

```python
isinstance(graph, ig.Graph)
```

```python
# number of nodes:
len(graph.vs)
```

```python
# number of edges:
len(graph.es)
```

Depending on the specific data source of the graph the graphs will have different node and edge attributes:

```python
node = list(graph.vs)[42]
```

```python
print(node.attributes())
```

```python
edge = list(graph.es)[0]
print(edge.attributes())
```

Usually there are always common gene identifiers, like gene symbol or entrez ID. Then there are attributes more specific to the data source at hand. In case of RegNetwork, which captures interactions of direct gene expression regulation between transcription factors and other genes and miRNAs and protein-coding genes, there is for example a node attribute 'node_type' indicating whether a gene encodes a miRNA , a transcription factor or any other protein.
